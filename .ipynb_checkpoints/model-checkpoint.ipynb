# Model JST Plant Climate Chamber

## Jaringan Saraf Tiruan untuk Plant Climate Chamber

```python
# Mengimpor pustaka yang dibutuhkan.

import numpy as np
import pandas as pd
from sklearn.neural_network import MLPRegressor
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Memasukkan dataset ke dalam jupyter.
data = pd.read_excel('Data/dataAll.xlsx')
```

```python
# Memisahkan data input dan data target.

# Data Input
X = data[['HT','AC','To','RD']]

# Data Target
target = data[['Td','RH','No','Variation','Month','Time']]
```

```python
# Memisahkan data pelatihan (training), data validasi (validation), dan data pengujian (testing).

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score, r2_score

def mape(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Memisahkan 20% menjadi data tes
X_temp, X_test, y_temp, y_test = train_test_split(X, target, test_size=0.05, shuffle=True, random_state=15)

# Memisahkan 20% menjadi data validation
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=(15/95), shuffle=True, random_state=2019)
```

```python
# Menghapus data temporary
del X_temp, y_temp

# print proportions
print('train: {}% | validation: {}% | test {}%'.format(round(len(y_train)/len(target)*100,2),
                                                       round(len(y_val)/len(target)*100,2),
                                                       round(len(y_test)/len(target)*100,2)))
```

```python
# banyak data
print(X_train.shape, X_val.shape, X_test.shape, y_train.shape, y_val.shape, y_test.shape)
```

### Model Tuning menggunakan Multi-Layer Perceptron Regressor

```python
# DESIGN PROCESS
print('DESIGN PROCESS')
print('-'*76)

f = open("Data/Model/designNN.txt","w+")

from fireTS.models import NARX, DirectAutoRegressor

# Perancangan Arsitektur
model_mlp = MLPRegressor(random_state = 1, hidden_layer_sizes=(91), max_iter = 5000)
# model_mlp = MLPRegressor(random_state = 1, activation = 'logistic', hidden_layer_sizes=(55), solver='lbfgs', 
#                          alpha=0.001, max_iter = 5000)

# Parameter Rancangan JST
random_state  = model_mlp.get_params()['random_state']
neuron_number = model_mlp.get_params()['hidden_layer_sizes']
activation    = model_mlp.get_params()['activation']
solver        = model_mlp.get_params()['solver']
max_iter      = model_mlp.get_params()['max_iter']
alpha         = model_mlp.get_params()['alpha']
learning_rate = model_mlp.get_params()['learning_rate']
shuffle       = model_mlp.get_params()['shuffle']

design = [neuron_number, activation, solver, max_iter, alpha, learning_rate, shuffle]

model_NARX = NARX(
    model_mlp,
    auto_order=2,
    exog_order=[2, 2],
    exog_delay=[1, 1])
        
f.write(str(design))
f.close()

print(design)
```

```python
# TRAINING PROCESS
print('TRAINING PROCESS')
print('-'*76)

# Melakukan pelatihan (training) menggunakan 'fit'
model_mlp.fit(X_train, y_train[['Td','RH']])
model_NARX.fit(X_train, y_train[['Td','RH']])

# Prediksi model
y_pred = model_mlp.predict(X_train)
y_pred = pd.DataFrame(y_pred, columns=['Td_pred','RH_pred'])
y_target = y_train[['Td','RH']]

# Membuat DataFrame hasil
y_target = y_target.reset_index()[['Td','RH']]
training = pd.concat([y_target,y_pred], axis=1)
training['e_Td'] = training.Td - training.Td_pred
training['e_RH'] = training.RH - training.RH_pred

# Menghitung kinerja
EVar = round(explained_variance_score(y_target, y_pred)*100,2)
R2   = round(r2_score(y_target, y_pred)*100,2)
RMSE = round(mean_squared_error(y_target, y_pred)**0.5,2)
MAE  = round(mean_absolute_error(y_target, y_pred),2)

# Menghitung nilai MAE, RMSE dan STDev
EVar_Td = round(explained_variance_score(y_train.Td, y_pred.Td_pred)*100,2)
RMSE_Td = round(mean_squared_error(y_train.Td, y_pred.Td_pred)**0.5,2)
MAE_Td  = round(mean_absolute_error(y_train.Td, y_pred.Td_pred),2)
Mean_Td = round(np.mean(training.e_Td),2)
Std_Td  = round(np.std(training.e_Td),2)

EVar_RH = round(explained_variance_score(y_train.RH, y_pred.RH_pred)*100,2)
RMSE_RH = round(mean_squared_error(y_train.RH, y_pred.RH_pred)**0.5,2)
MAE_RH  = round(mean_absolute_error(y_train.RH, y_pred.RH_pred),2)
Mean_RH = round(np.mean(training.e_RH),2)
Std_RH  = round(np.std(training.e_RH),2)

# print errors as report
print('EVar = {}% R2 = {}% RMSE = {} MAE = {}'.format(EVar, R2, RMSE, MAE))
print()
print('Td : EVar = {}% RMSE = {} MAE = {} Mean = {} Std = {}'.format(EVar_Td, RMSE_Td, MAE_Td, Mean_Td, Std_Td))
print('RH : EVar = {}% RMSE = {} MAE = {} Mean = {} Std = {}'.format(EVar_RH, RMSE_RH, MAE_RH, Mean_RH, Std_RH))
print()

# Menyimpan data dalam berkas
training.to_excel("Data/Model/Training.xlsx")
```

```python
print('VALIDATION PROCESS')
print('-'*76)

# Prediksi model
y_pred = model_mlp.predict(X_val)
y_pred = pd.DataFrame(y_pred, columns=['Td_pred','RH_pred'])
y_target = y_val[['Td','RH']]

# Membuat DataFrame hasil
y_target = y_target.reset_index()[['Td','RH']]
validation = pd.concat([y_target,y_pred], axis=1)
validation['e_Td'] = validation.Td - validation.Td_pred
validation['e_RH'] = validation.RH - validation.RH_pred

# Menghitung kinerja
EVar = round(explained_variance_score(y_target , y_pred)*100,2)
R2   = round(r2_score(y_target , y_pred)*100,2)
RMSE = round(mean_squared_error(y_target, y_pred)**0.5,2)
MAE  = round(mean_absolute_error(y_target , y_pred),2)

# Menghitung nilai MAE, RMSE dan STDev
MAE_Td  = round(mean_absolute_error(y_val.Td, y_pred.Td_pred),2)
RMSE_Td = round(mean_squared_error(y_val.Td, y_pred.Td_pred)**0.5,2)
EVar_Td = round(explained_variance_score(y_val.Td, y_pred.Td_pred)*100,2)
Mean_Td = round(np.mean(validation.e_Td),2)
Std_Td  = round(np.std(validation.e_Td),2)

MAE_RH  = round(mean_absolute_error(y_val.RH, y_pred.RH_pred),2)
RMSE_RH = round(mean_squared_error(y_val.RH, y_pred.RH_pred)**0.5,2)
EVar_RH = round(explained_variance_score(y_val.RH, y_pred.RH_pred)*100,2)
Mean_RH = round(np.mean(validation.e_RH),2)
Std_RH  = round(np.std(validation.e_RH),2)

# print errors as report
print('EVar = {}% R2 = {}% RMSE = {} MAE = {}'.format(EVar, R2, RMSE, MAE))
print()
print('Td : EVar = {}% RMSE = {} MAE = {} Mean = {} Std = {}'.format(EVar_Td, RMSE_Td, MAE_Td, Mean_Td, Std_Td))
print('RH : EVar = {}% RMSE = {} MAE = {} Mean = {} Std = {}'.format(EVar_RH, RMSE_RH, MAE_RH, Mean_RH, Std_RH))
print()

# Menyimpan data dalam berkas
validation.to_excel("Data/Model/Validation.xlsx")
```

```python
# TESTING PROCESS
print('TESTING PROCESS')
print('-'*76)

# Prediksi model
y_pred = model_mlp.predict(X_test)
y_pred = pd.DataFrame(y_pred, columns=['Td_pred','RH_pred'])
y_target = y_test[['Td','RH']]

#Membuat DataFrame hasil
y_target = y_target.reset_index()[['Td','RH']]
testing = pd.concat([y_target,y_pred], axis=1)
testing['e_Td'] = testing.Td - testing.Td_pred
testing['e_RH'] = testing.RH - testing.RH_pred

# Menghitung kinerja
EVar = round(explained_variance_score(y_target, y_pred)*100,2)
R2   = round(r2_score(y_target, y_pred)*100,2)
RMSE = round(mean_squared_error(y_target, y_pred)**0.5,2)
MAE  = round(mean_absolute_error(y_target, y_pred),2)

# Menghitung nilai MAE, RMSE dan STDev
EVar_Td = round(explained_variance_score(y_test.Td, y_pred.Td_pred)*100,2)
MAE_Td  = round(mean_absolute_error(y_test.Td, y_pred.Td_pred),2)
RMSE_Td = round(mean_squared_error(y_test.Td, y_pred.Td_pred)**0.5,2)
Mean_Td = round(np.mean(testing.e_Td),2)
Std_Td  = round(np.std(testing.e_Td),2)

MAE_RH  = round(mean_absolute_error(y_test.RH, y_pred.RH_pred),2)
RMSE_RH = round(mean_squared_error(y_test.RH, y_pred.RH_pred)**0.5,2)
EVar_RH = round(explained_variance_score(y_test.RH, y_pred.RH_pred)*100,2)
Mean_RH = round(np.mean(testing.e_RH),2)
Std_RH  = round(np.std(testing.e_RH),2)

# print errors as report
print('EVar = {}% R2 = {}% RMSE = {} MAE = {}'.format(EVar, R2, RMSE, MAE))
print()
print('Td : EVar = {}% RMSE = {} MAE = {} Mean = {} Std = {}'.format(EVar_Td, RMSE_Td, MAE_Td, Mean_Td, Std_Td))
print('RH : EVar = {}% RMSE = {} MAE = {} Mean = {} Std = {}'.format(EVar_RH, RMSE_RH, MAE_RH, Mean_RH, Std_RH))
print()

# Menyimpan data dalam berkas
testing.to_excel("Data/Model/Testing.xlsx")
```

```python
i = 0
for v in y_test.Variation.unique():
    print('Variasi:', v)
    
    x1  = y_test.loc[y_test['Variation']==v].Time
    
    y1  = X_test.loc[y_test['Variation']==v].AC
    y2  = X_test.loc[y_test['Variation']==v].HT
    y3  = X_test.loc[y_test['Variation']==v].To
    y4  = X_test.loc[y_test['Variation']==v].RD
    y5a = y_test.loc[y_test['Variation']==v].Td
    y5b = y_pred.loc[y_pred['Variation']==v].Td_pred
    y6a = y_test.loc[y_test['Variation']==v].RH
    y6b = y_pred.loc[y_pred['Variation']==v].RH_pred
    
    plt.subplots(figsize=(30,30))
    
    plt.subplot(3,2,1)
    plt.title('Grafik Nilai-Nilai Variabel terhadap waktu. Variasi: ' + str(v))
    plt.plot(x1, y3, 'o')
    plt.xlabel(None)
    plt.ylabel('To')
    
    plt.subplot(3,2,2)
    plt.plot(x1, y4, 'o')
    plt.xlabel(None)
    plt.ylabel('Radiation')
    
    plt.subplot(3,2,3)
    plt.plot(x1, y1, 'o')
    plt.xlabel(None)
    plt.ylabel('AC')
    
    plt.subplot(3,2,4)
    plt.plot(x1, y2, 'o')
    plt.xlabel(None)
    plt.ylabel('Heater')
    
    plt.subplot(3,2,5)
    plt.plot(x1, y5a, 'o', label='target')
    plt.plot(x1, y5b, 'ro', label='prediction')
    plt.xlabel(None)
    plt.ylabel('Td')
    plt.legend(loc = 'upper left')
    
    plt.subplot(3,2,6)
    plt.plot(x1, y6a, 'o', label='target')
    plt.plot(x1, y6b, 'ro', label='prediksi')
    plt.xlabel('waktu')
    plt.ylabel('RH')
    plt.legend(loc = 'upper left')
    
    plt.show()
    
    print('_'*110)
    print()
    
    i+=1
    if i == 3:
        break
```
